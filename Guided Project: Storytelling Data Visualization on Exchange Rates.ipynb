# Storytelling Data Visualization on Exchange Rates

## Introduction

In this project, we'll be working on visualizing the daily exchange rates dataset of the Euro since 1999. The dataset was compiled by the European Central Bank and it's available on their website. The dataset is named `euro-daily-hist_1999_2022.csv`.

Let's get a sense of our data and what we're working with. We'll start by reading the data into a pandas DataFrame and then printing the first and last five rows of the data.

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

```python
exchange_rates.describe()
```

## Data Cleaning

Our main focus of this project is on the exchange rate between the Euro and the American Dollar. We will filter the data to only include the exchange rate between these two currencies. We will also rename the columns to make them more descriptive and easier to work with. We will then convert the `Time` column to a datetime data type and sort the data by the `Time` column.

```python
exchange_rates.rename(columns={'Period\\Unit:': 'Time', '[US dollar ]': 'US_dollar'}, inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.head()
```

```python
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

Now's lets analyse the euro-dollar data to inspect for missing values and decide on how to handle them.

```python
euro_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_dollar['US_dollar'].value_counts()
```

Let's create a new DataFrame that doesn't contain any missing values in the `US_dollar` column and convert the `US_dollar` column to a float data type.

```python
euro_dollar = euro_dollar[euro_dollar['US_dollar'] != '-']
euro_dollar['US_dollar'] = euro_dollar['US_dollar'].astype(float)
euro_dollar.info()
```

## Rolling Mean

We will calculate the rolling mean for the exchange rate between the Euro and the American Dollar to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(euro_dollar['Time'], euro_dollar['US_dollar'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.subplot(3, 2, 1)
plt.plot(euro_dollar['Time'], euro_dollar['US_dollar'])
plt.title('Original Values', weight='bold')

for i, rolling_mean in zip(range(2, 7), [7, 30, 50, 100, 365]):
    plt.subplot(3, 2, i)
    plt.plot(euro_dollar['Time'], euro_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title(f'Rolling Window: {rolling_mean}', weight='bold')

plt.tight_layout() #Automatically adjusts the subplots paddingto give a better layout
plt.show()
```

```python
euro_dollar['rolling_mean'] = euro_dollar['US_dollar'].rolling(30).mean()
euro_dollar 
```

## 1. Financial Crisis (2007-2008)

```python
financial_crisis = euro_dollar.copy()[(euro_dollar['Time'].dt.year >= 2006) & (euro_dollar['Time'].dt.year <= 2009)]
financial_crisis_07_08 = euro_dollar.copy()[(euro_dollar.Time.dt.year >= 2007) & (euro_dollar.Time.dt.year <= 2008)]
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
### Creating the plot
fig, ax =plt.subplots(figsize=(12, 6))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=1, color='#A6D785')

### Highlighting the 2007-2008 period
ax.plot(financial_crisis_07_08['Time'], financial_crisis_07_08['rolling_mean'], linewidth=3, color='#e23d28')

### Highlihting the peak of the crisis
ax.axvspan(xmin=13950, xmax=14140, ymin=0.09, alpha=0.3, color='grey')

### Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 13110
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x=x, y=1.13, s=year, alpha=0.5, fontsize=11)
    x += 365

y = 1.193
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(x=13020, y=y, s=rate, alpha=0.5, fontsize=11)
    y += 0.1

### Adding a title and a subtitle
ax.text(x=13020, y=1.67, s="Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight='bold', alpha=0.75)
ax.text(x=13020, y=1.63, s="Euro-USD exchange rates between 2006 and 2010", size=12)

### Adding a signature
ax.text(x=13020, y=1.07, s='Â©LIMEWORK TECH' + ' '*170 + 'Source: European Central Bank', color='#f0f0f0', backgroundcolor='#4d4d4d', size=10)

### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```

## 2. The last three US presidents

Let's start this visualization by creating a new column that categorizes the `Time` column into the three presidencies. We will then calculate the mean exchange rate for each presidency and visualize the results.

```python
bush_obama_trump = euro_dollar.copy()[(euro_dollar['Time'].dt.year >= 2001) & (euro_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

Below, we'll use matplotlib functional approach to build the graphs. This appraoch offers more flexibility and control over the plots. First we will:
- build 3 of the graphs on a 2x3 grid(this grid should have six subplots, but we only build three; the bottom row remains empty).
- We will then build only the bottom graph of a 2x1 grid (this grid should have two subplots).
- The two grids will be combined, and we'll end up with three graph on the top row and one graph on the bottom row.

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Creating the subplots
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 1, 2)

axes = [ax1, ax2, ax3, ax4]

### Changes to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.3)
    ax.grid(alpha=0.5)

### AX Plot 1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'], color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3)
ax1.text(x=12300, y=1.92, s="BUSH", weight='bold', color='#BF5FFF', fontsize=18)
ax1.text(x=12000, y=1.8, s="(2001-2009)", weight='bold', alpha=0.3)


### AX Plot 2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'], color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha=0.3)
ax2.text(x=15000, y=1.92, s="OBAMA", weight='bold', color='#ffa500', fontsize=18)
ax2.text(x=14840, y=1.8, s="(2009-2017)", weight='bold', alpha=0.3)

### AX Plot 3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'], color='#00B2EE')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha=0.3)
ax3.text(x=17550, y=1.92, s="TRUMP", weight='bold', color='#00B2EE', fontsize=18)
ax3.text(x=17450, y=1.8, s="(2017-2021)", weight='bold', alpha=0.3)

### AX Plot 4: All Presidents
ax4.plot(bush['Time'], bush['rolling_mean'], color='#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#00B2EE')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(x=11600, y=2.3, s="EURO-USD rate averaged 1.22 under the last three US presidents", weight='bold', alpha=0.75, size=18)
ax1.text(x=11000, y=2.2, s="EURO-USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)", size=12)

### Adding a signature
ax4.text(x=10600, y=0.65, s='Â©LIMEWORK TECH' + ' '*180 + 'Source: European Central Bank', color='#f0f0f0', backgroundcolor='#4d4d4d', size=10)
plt.show()
```

## 3. The pandemic period

Lets start by defining the `Time` column into two periods: the pandemic period (2020) and the baseline period (2016-2019). We will then compare exchange rate for each period and visualize the results.

```python
pre_pandemic = euro_dollar.copy()[(euro_dollar['Time'].dt.year >= 2016) & (euro_dollar['Time'].dt.year <= 2020)]
pandemic = euro_dollar.copy()[euro_dollar['Time'].dt.year == 2020]
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Creating the plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(pre_pandemic['Time'], pre_pandemic['rolling_mean'], linewidth=3, color='#A6D785')

### Highlighting the 2020 period
ax.plot(pandemic['Time'], pandemic['rolling_mean'], linewidth=3, color='#e23d28')

### Annotating the two periods using arrows
ax.annotate('',
            xy=(pd.Timestamp('2016-01-01'), 1.05),
            xytext=(pd.Timestamp('2019-12-31'), 1.05),
            xycoords='data',
            arrowprops=dict(arrowstyle='<->', color='black',linewidth=2)
           )

ax.annotate('',
            xy=(pd.Timestamp('2020-01-01'), 1.075),
            xytext=(pd.Timestamp('2020-12-31'), 1.075),
            xycoords='data',
            arrowprops=dict(arrowstyle='<->', color='black', linewidth=2)
           )

###Highlighting the peak of the pandemic
ax.axvspan(xmin=17790, xmax=17485, ymin=0.034, alpha=0.3, color='grey')
ax.axvspan(xmin=18630, xmax=18260, ymin=0.034, alpha=0.3, color='grey')

### Adding the title and subtitle
ax.text(x=16600, y=1.25, s="EURO-USD rate was peaked at about 1.250 before the 2020 pandemic", weight='bold', alpha=0.75)
ax.text(x=16600, y=1.24, s="EURO-USD exchange rates between 2016 and 2020", size=12)

### Adding a signature
ax.text(x=16600, y=1.02, s='Â©LIMEWORK TECH' + ' '*171 + 'Source: European Central Bank', color='#f0f0f0', backgroundcolor='#4d4d4d', size=10)

### Add some transparency to the grid
ax.grid(alpha=0.5)


plt.show()
```
