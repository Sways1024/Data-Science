# Euro_Dollar Data Visualization Project

## Introduction

In this project, we'll be working on visualizing the daily exchange rates dataset of the Euro since 1999. The dataset was compiled by the European Central Bank and it's available on their website. We'll be analyzing the daily exchange rate of the euro against various currencies from 1999 to 2021. The euro, denoted by the symbol €, serves as the primary currency for most countries within the European Union.

To understand how exchange rates function, consider this example: if the exchange rate from the euro to the US dollar is 1.25, it means you would get $1.25 for every €1. This implies that the euro is stronger than the US dollar in this scenario.

The dataset was sourced from the European Central Bank. While, this dataset is regularly updated, and for our purposes, we will use the data as of January 2024. The dataset is named `euro-daily-hist_1999_2022.csv`. We will specifically examine the exchange rate trends between the euro and the US dollar. 

To get sense of the data we're working with, we'll start by reading the data into a pandas DataFrame and then printing the first and last five rows of the data.

```python
#Import the pandas library
import pandas as pd

#Read the data set
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv')
```

## Data Cleaning

The focus of this project is on the exchange rate between the Euro and the American Dollar(the Eur-Usd pair). To clean the data, we'll filter the data to only include the exchange rate between these two currencies we're interested in. Next, we will rename the columns to make them more descriptive and easier to work with. The final part of the cleaning will require us to rename the `Period\\Unit:` to `Time`, converting the new column to a datetime data type using the pd.to_datetime method, sorting final result by the `Time` column in ascending order and resetting the index while dropping the initial one.

```python
#Renaming and replacing the columns 
exchange_rates.rename(columns={'Period\\Unit:': 'Time', '[US dollar ]': 'US_dollar'}, inplace=True)

#Converting the Time column to a datetime data type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

#Sorting the data by the Time column in ascending order
exchange_rates.sort_values('Time', inplace=True)

#Resetting the index
exchange_rates.reset_index(drop=True, inplace=True)

#Display the first five rows of the cleaned dataset
exchange_rates.head()
```

As part of the cleaning process, we will create a new dataframe by isolating the the `Time` and `US_dollar`. Furthermore, we will then analyse this euro-dollar data to inspect for missing values and decide on how to handle them.

```python
#Creating a new DataFrame to contain the Time and US_dollar columns
euro_dollar = exchange_rates[['Time', 'US_dollar']].copy()

#Checking for missing values using the value_counts method
euro_dollar['US_dollar'].value_counts()
```

We can see that the dataframe contains 62 '-' values, which may represent missing values. This missing values only make up a small portion of our data. To handle that, we will replace the `euro_dollar` new DataFrame that doesn't contain any missing values in the `US_dollar` column and convert the `US_dollar` column to a float data type.

```python
#Remove the rows with the '-' values
euro_dollar = euro_dollar[euro_dollar['US_dollar'] != '-']

#Convert the US_dollar column to a float data type
euro_dollar['US_dollar'] = euro_dollar['US_dollar'].astype(float)

#Display the summary information after cleaning the data conversion
euro_dollar.info()
```

## Rolling Mean

For the final part of the analysis, we will generate the rolling mean for the exchange rate between the Euro and the American Dollar to visualize the evolution of the euro-dollar exchange rate from 1999 to 2024.

```python
#import the matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline 


# Set the plot style and figure size
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8, 4))

#Plot the euro-dollar exchange rate
plt.plot(euro_dollar['Time'], euro_dollar['US_dollar'], linewidth=1, color='#FF4500')

#Add a title and a subtitle
plt.title('Euro-USD Exchange Rate From 1999-2024', weight='bold', size=12)

#Display the plot
plt.show()
```

When examining the plot line, we observe numerous small fluctuations rather than a seamless, smooth line. These fluctuations represent the day-to-day changes in the exchange rate, demonstrating how it can rise and fall on a daily basis. 

For our analyses, it will be beneficial to minimize the visibility of these daily fluctuations to better focus on the overarching trends. This can be achieved through the use of a rolling mean. The term "rolling window" or "moving window" refers to the number of days considered when calculating the rolling mean. For instance, with a rolling window of 3 days, the rolling mean is computed using the exchange rate from the current day and the two preceding days. Consequently, on the third day, the moving average is derived from the exchange rates of the first three days.

```python
#Set the plot style and figure size
plt.figure(figsize=(9,8))
plt.subplot(3, 2, 1)

#Main graph title
plt.suptitle('Rolling Mean for the various windows indicating graph smoothening', weight='bold', size=16)

#Plot the original values of the euro-dollar exchange rate
plt.plot(euro_dollar['Time'], euro_dollar['US_dollar'], linewidth=1)

#Add a title
plt.title('Original Values', weight='bold', size=14)

#Create a for loop to generate the rolling mean
for i, rolling_mean in zip(range(2, 7), [7, 30, 50, 100, 365]):
    plt.subplot(3, 2, i)
    plt.plot(euro_dollar['Time'], euro_dollar['US_dollar'].rolling(rolling_mean).mean(), linewidth=1)

    plt.title(f'Rolling Window: {rolling_mean}', weight='bold', size=14)

#Automatically adjusts the subplots paddingto give a better layout
plt.tight_layout() 

#Display the plot
plt.show()
```

The above shows the rolling mean for the various windows indicating graph smoothening. The first graph shows the original values of the euro-dollar exchange rate. The subsequent graphs show the rolling mean for the following windows: 7, 30, 50, 100, and 365.

Below, we will calculate the rolling mean on the `euro_dollar` DataFrame for the 30 day period and visualize the results.

```python
#Rolling window for the 30-day period
euro_dollar['rolling_mean'] = euro_dollar['US_dollar'].rolling(30).mean()
euro_dollar 
```

## 2. The last three US presidents

Let's start this visualization by spliiting the `euro_dollar` dataframe into the three timelines for each president. We will then calculate the mean exchange rate for each presidency and visualize the results.

```python
#Select data for the entire period of the last three US presidents
bush_obama_trump = euro_dollar.copy()[(euro_dollar['Time'].dt.year >= 2001) & (euro_dollar['Time'].dt.year < 2021)]

#Select data for each president
bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

Below, we'll use matplotlib functional approach to build the graphs. This approach offers more flexibility and control over the plots. First we will:
- Build 3 of the graphs on a 2x3 grid (this grid should have six subplots, but we only build three; the bottom row remains empty).
- We will then build only the bottom graph of a 2x1 grid (this grid should have two subplots).
- The two grids will be combined, and we'll end up with three graph on the top row and one graph on the bottom row.

```python
#Add the FiveThirtyEight style
style.use('fivethirtyeight')

#Create the subplots
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 1, 2)

axes = [ax1, ax2, ax3, ax4]

#Make changes to all the y axis labels and gridlines in the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.3)
    ax.grid(alpha=0.5)
    

#AX Plot 1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'], color='#BF5FFF', linewidth=2)
ax1.set_xticks(['2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009'])
ax1.set_xticklabels(['2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3)
ax1.text(x=12300, y=1.92, s="BUSH", weight='bold', color='#BF5FFF', fontsize=18)
ax1.text(x=12000, y=1.8, s="(2001-2009)", weight='bold', alpha=0.3)


### AX Plot 2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'], color='#ffa500', linewidth=2)
ax2.set_xticks(['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
ax2.set_xticklabels(['2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha=0.3)
ax2.text(x=15000, y=1.92, s="OBAMA", weight='bold', color='#ffa500', fontsize=18)
ax2.text(x=14840, y=1.8, s="(2009-2017)", weight='bold', alpha=0.3)

#X Plot 3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'], color='#00B2EE', linewidth=2)
ax3.set_xticks(['2017','2017-07', '2018','2018-07', '2019', '2019-07', '2020','2020-07', '2021'])
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha=0.3)
ax3.text(x=17550, y=1.92, s="TRUMP", weight='bold', color='#00B2EE', fontsize=18)
ax3.text(x=17450, y=1.8, s="(2017-2021)", weight='bold', alpha=0.3)

#AX Plot 4: All Presidents
ax4.plot(bush['Time'], bush['rolling_mean'], color='#BF5FFF', linewidth=2)
ax4.plot(obama['Time'], obama['rolling_mean'], color='#ffa500', linewidth=2)
ax4.plot(trump['Time'], trump['rolling_mean'], color='#00B2EE', linewidth=2)
ax4.grid(alpha=0.5)

#Add a title and a subtitle
ax1.text(x=11600, y=2.3, s="EURO-USD rate averaged 1.22 under the last three US presidents", weight='bold', alpha=0.75, size=18)
ax1.text(x=11000, y=2.2, s="EURO-USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)", size=12)

#Add a signature
ax4.text(x=10600, y=0.65, s='©LIMEWORK TECH' + ' '*180 + 'Source: European Central Bank', color='#f0f0f0', backgroundcolor='#4d4d4d', size=10)
plt.show()
```

# 3. The pandemic period

Lets start by splitting the `euro_dollar` dataframe into three time periods: The pre-pandemic(2016-2019), the pandemic(2019) and post-pandemic periods(2021-2024). We will then compare exchange rate for each period and visualize the results.

```python
#Select the data for each time period
pre_pandemic = euro_dollar.copy()[(euro_dollar['Time'].dt.year >= 2016) & (euro_dollar['Time'].dt.year <= 2020)]
pandemic = euro_dollar.copy()[euro_dollar['Time'].dt.year == 2020]
post_pandemic = euro_dollar.copy()[euro_dollar['Time'].dt.year > 2020]
```

```python
#Add the FiveThirtyEight style
style.use('fivethirtyeight')

#Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

#Plot the pre-pandemic period graph
ax.plot(pre_pandemic['Time'], pre_pandemic['rolling_mean'], linewidth=1, color='#A6D785')

#Plot the pandemic period graph
ax.plot(pandemic['Time'], pandemic['rolling_mean'], linewidth=3, color='#e23d28')

#Plot the post-pandemic period graph
ax.plot(post_pandemic['Time'], post_pandemic['rolling_mean'], linewidth=1, color='#A6D785')

#Deliniating the pre-pandemic periods using an arrow
ax.annotate('',
            xy=(pd.Timestamp('2016-01-01'), 1.04),
            xytext=(pd.Timestamp('2019-12-31'), 1.04),
            xycoords='data',
            arrowprops=dict(arrowstyle='<->', color='black',linewidth=2)
           )

#Deliniate the pandemic periods using an arrow
ax.annotate('',
            xy=(pd.Timestamp('2020-01-01'), 1.075),
            xytext=(pd.Timestamp('2020-12-31'), 1.075),
            xycoords='data',
            arrowprops=dict(arrowstyle='<->', color='black', linewidth=2)
           )

#Deliniate the post-pandemic periods using an arrow
ax.annotate('',
            xy=(pd.Timestamp('2021-01-01'), 1.0),
            xytext=(pd.Timestamp('2024-01-01'), 1.0),
            xycoords='data',
            arrowprops=dict(arrowstyle='<->', color='black', linewidth=2)
           )

#Highlight the peak of the pandemic
ax.axvspan('2020', '2021', ymin=0.3, ymax=0.92, alpha=0.3, color='grey')

#Add the title and subtitle
ax.text(x=16800, y=1.26, s="EURO-USD rate was peaked at about 1.250 before the 2020 pandemic", weight='bold', alpha=0.75)
ax.text(x=17200, y=1.24, s="EURO-USD exchange rates before, during and after the pandemic", size=12)

#Add a signature
ax.text(x=16100, y=0.92, s='©LIMEWORK TECH' + ' '*170 + 'Source: European Central Bank', color='#f0f0f0', backgroundcolor='#4d4d4d', size=10)

#Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```
